```python
import numpy as np
import matplotlib.pyplot as plt

from terrain import DiscreteTerrain
from prospect import UniformProspect, EGreedyProspect, SoftmaxProspect, UCB1Prospect
from vizualizer import Vizualizer
```

```python
terrain = DiscreteTerrain(10, 10)
```

```python
np.max(terrain.grid)
```

### Uniform

```python
prospect = UniformProspect(trials=1000, terrain=terrain)
prospect.prospect()
viz = Vizualizer(prospect.knowledge)
viz.graph_2D(prospect)
```

### Epsilon-Greedy

```python
prospect = EGreedyProspect(trials=1000, terrain=terrain)
prospect.prospect()
viz = Vizualizer(prospect.knowledge)
viz.graph_2D(prospect)
```

### Softmax

```python
prospect = SoftmaxProspect(trials=1000, terrain=terrain)
prospect.prospect()
viz = Vizualizer(prospect.knowledge)
viz.graph_2D(prospect)
```

### UCB1

```python
prospect = UCB1Prospect(trials=1000, terrain=terrain)
prospect.prospect()
viz = Vizualizer(prospect.knowledge)
viz.graph_2D(prospect)
```

### All

```python
for prospect_class in [UniformProspect, EGreedyProspect, SoftmaxProspect, UCB1Prospect]:
    prospect = prospect_class(trials=1000, terrain=terrain)
    prospect.prospect()
    viz = Vizualizer(prospect.knowledge)
    viz.graph_2D(prospect)
plt.legend(['UniformProspect', 'EGreedyProspect', 'SoftmaxProspect', 'UCB1Prospect'])
plt.show()
```
